```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_excel("../dataset/dataset.xlsx")
```

```python
# Display basic information about the dataset
print("Dataset Information:")
print(df.info())
```

```python
# Display the first few rows of the dataset
print("\nFirst Few Rows of the Dataset:")
print(df.head())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x="generated", data=df)
plt.title("Distribution of Target Variable")
plt.show()
```

```python
# Word cloud for the text column
from wordcloud import WordCloud
```

```python
# Convert the "text" column to strings
df["text"] = df["text"].astype(str)

# Generate word cloud
text_wordcloud = WordCloud(width=800, height=400, background_color="white").generate(" ".join(df["text"]))

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(text_wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud for Text Column")
plt.show()
```
