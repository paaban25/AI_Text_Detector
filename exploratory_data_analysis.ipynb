```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("../dataset/dataset.xlsx")
```

```python
print("Dataset Information:")
print(df.info())
```

```python
print("\nFirst Few Rows of the Dataset:")
print(df.head())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x="generated", data=df)
plt.title("Distribution of Target Variable")
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
df["text"] = df["text"].astype(str)

text_wordcloud = WordCloud(width=800, height=400, background_color="white").generate(" ".join(df["text"]))

plt.figure(figsize=(10, 5))
plt.imshow(text_wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud for Text Column")
plt.show()
```
